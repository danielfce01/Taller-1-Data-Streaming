```
import json
import yaml
from dicttoxml import dicttoxml
from xml.dom.minidom import parseString

# --- JSON ---
json_data = json.dumps(ventas_garantias, indent=4, ensure_ascii=False)
with open("local/data/ventas_garantias.json", "w", encoding="utf-8") as f:
    f.write(json_data)

# --- XML con formato legible ---
xml_bytes = dicttoxml(ventas_garantias, custom_root="VentasGarantias", attr_type=False)
dom = parseString(xml_bytes)  # convierte los bytes a DOM
xml_pretty = dom.toprettyxml(indent="  ")  # agrega indentación

with open("local/data/ventas_garantias.xml", "w", encoding="utf-8") as f:
    f.write(xml_pretty)

# --- YAML ---
yaml_data = yaml.dump(ventas_garantias, allow_unicode=True, sort_keys=False)
with open("local/data/ventas_garantias.yaml", "w", encoding="utf-8") as f:
    f.write(yaml_data)

# --- Impresión en consola ---
print("=== JSON ===")
print(json_data)

print("\n=== XML ===")
print(xml_pretty)

print("\n=== YAML ===")
print(yaml_data)
```

#**Propuesta para la representación de los datos**

Teniendo en cuenta las actividades realizadas en los putnos anteriores, proponga cual considera que es la representación adecuada par estos datos (Defina los criterios que considera mas relevantes para su decisión)

**Criterios**

   - Rapidez de transmisión:

JSON: Es un formato ligero y rápido para serializar/deserializar, pero puede ser más lento que los binarios.

XML: Es más verboso y por lo tanto más lento en transmisión y procesamiento.

YAML: Similar a JSON en legibilidad, pero más lento debido a la necesidad de analizar la indentación y caracteres especiales.

Protocol Buffers: Binario, muy eficiente en tamaño y velocidad. Diseñado para ser rápido.

Apache Thrift: Similar a Protocol Buffers, binario y eficiente.

 - Costo:

JSON, XML, YAML: Son estándares abiertos y no tienen costo de licencia.

Protocol Buffers: Open source, sin costo.

Apache Thrift: Open source, sin costo.

 - Seguridad:

 Los formatos binarios (Protocol Buffers, Thrift) son más difíciles de manipular directamente, lo que puede ser una ventaja. Pero la seguridad depende más del canal de transmisión y el cifrado.

 -  Recursos necesarios:

JSON, XML, YAML: Requieren más ancho de banda debido a su verbosidad (especialmente XML y YAML). También requieren más CPU para analizar.

Protocol Buffers y Thrift: Son binarios, por lo que usan menos ancho de banda y son más eficientes en CPU.

 - Capacidad de recuperar la información:

JSON, XML, YAML: Son formatos autodescriptivos y fáciles de leer, por lo que la recuperación es sencilla.

Protocol Buffers y Thrift: Requieren el esquema para decodificar, lo que puede ser un punto débil si se pierde el esquema. Pero con el esquema, la recuperación es completa.

 - Estabilidad frente a fallas:

JSON, XML, YAML: Son formatos maduros y ampliamente soportados. Cualquier sistema puede leerlos.

Protocol Buffers y Thrift: Son estables, pero si hay cambios en el esquema, se debe manejar la compatibilidad.

- Recomendación:

Protocol Buffers, por su eficiencia, bajo consumo de recursos y amplia adopción. Sin embargo, si la legibilidad es importante, JSON es aceptable pero menos eficiente
